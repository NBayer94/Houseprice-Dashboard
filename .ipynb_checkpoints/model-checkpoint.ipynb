## Preprocessing

```python
X = train.drop(columns=['SalePrice'])
y = train['SalePrice']
log_y = np.log(y)
```

```python
X.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
num_pipe = Pipeline(steps=[('imputer', KNNImputer()),
                           ('scaler', MinMaxScaler())])

cat_pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                           ('encoder', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(transformers=[('num', num_pipe, num_cols),
                                                     ('cat', cat_pipe, cat_cols)])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_log_error, make_scorer
```

```python
inner_cv = KFold(shuffle=True, random_state=42)
outer_cv = KFold(shuffle=True, random_state=42)

param_grid_rfr = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 2 , 4, 6],
    'classifier__min_samples_split': [2, 4, 6]
}

param_grid_gbr = {
    'classifier__learning_rate': [0.1, 0.2, 0.4],
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 3 , 6],
    'classifier__min_samples_split': [2, 4, 6]
}

def rmlse(y_true, y_pred, **kwargs):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))

def neg_rmlse(y_true, y_pred, **kwargs):
    return - np.sqrt(mean_squared_log_error(y_true, y_pred))

neg_rmlse_score = make_scorer(neg_rmlse)
```

```python
# Nested Gridsearh GBR

gbr = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', GradientBoostingRegressor(random_state=42))])

#gsCV = GridSearchCV(gbr, param_grid=param_grid_gbr, scoring=neg_rmlse_score, cv=inner_cv, n_jobs=-1)
#cv_score_gbr = cross_val_score(gsCV, X, y, scoring=neg_rmlse_score, cv=outer_cv, n_jobs=-1)
```

```python
gsCV = GridSearchCV(gbr, param_grid=param_grid_gbr, scoring=neg_rmlse_score, cv=inner_cv, n_jobs=-1)
gsCV.fit(X,y)
```

```python
gsCV.best_params_
```

```python
gbr.set_params(**gsCV.best_params_)
gbr.fit(X, y)
```

```python
y_pred = gbr.predict(X)
```

```python
rmlse(y, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

### Random Forest

```python
# Nested Gridsearh GBR

rfr = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestRegressor(random_state=42, n_jobs=-1))])

gsCV_rf = GridSearchCV(rfr, param_grid=param_grid_rfr, scoring=neg_rmlse_score, cv=inner_cv, n_jobs=-1)
gsCV_rf.fit(X,y)
```

```python
rfr.set_params(**gsCV_rf.best_params_)
rfr.fit(X, y)
y_pred_rfr = rfr.predict(X)
```

```python
rmlse(y_pred_rfr, y_pred)
```

```python
r2_score(y_pred_rfr, y_pred)
```

### XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgbr = Pipeline(steps=[('preprocessor', preprocessor),
                       ('model', XGBRegressor(random_state=42, n_jobs=-1))])

param_grid_xgbr = {
    'model__n_estimators': [50, 100, 200],
    'model__learning_rate': [0.1, 0.05, 0.15],
    'model__max_depth': [4, 6, 8],
    'model__subsample': [1, 0.75, 0.5],
    'model__colsample_bytree': [1, 0.75, 0.5],
    'model__reg_lambda': [1, 0, 10],
    'model__gamma': [0, 0.25, 1]
}
```

```python
gsCV_xgbr = GridSearchCV(xgbr, param_grid=param_grid_xgbr, scoring=neg_rmlse_score, cv=inner_cv, n_jobs=-1)
gsCV_xgbr.fit(X,y)
```

```python
xgbr.set_params(**gsCV_xgbr.best_params_)
xgbr.fit(X, y)
y_pred_xgbr = xgbr.predict(X)
```
